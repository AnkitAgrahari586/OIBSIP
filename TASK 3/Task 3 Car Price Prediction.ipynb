**Importing necessary Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading Dataset**

```python
df=pd.read_excel('car price.xlsx')
df
```

**Information About Dataset**

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()*100
```

```python
df=df.drop(columns=["car_ID",'CarName','enginetype','cylindernumber','fueltype','aspiration','fuelsystem','doornumber','carbody','drivewheel','enginelocation'])
df
```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df['price']
y
```

**Loading The Model**

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score as r
```

```python
r(y_test,ypred)*100
```
